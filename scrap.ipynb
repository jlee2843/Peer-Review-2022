```python
import json
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
for cursor in range(10):
    url = f"https://api.biorxiv.org/pubs/biorxiv/2000-03-01/2020-03-30/{cursor*100+1}"
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urlopen(req).read()
    soup = BeautifulSoup(webpage)
    with open(f"data-{cursor}.json",'w',encoding='utf-8') as f:
        f.write(soup.text)
```

```python
soup.prettify()
```

```python
i = 4
try:
    with open(f"data-{i}.json") as f:
        data = json.load(f)
except Exception as e:
    print(e)
    with open(f"data-{i}.json", 'r') as f:
        errIndex = int(str(e).split()[-1][:-1])
        print(f.read()[errIndex-2:errIndex+2])
```
